```python
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
name = "YS01 _ZnSe_in_MilliQ_2_13_2024.csv"
#name = 'ZnSe_bulk_solution_in_hexanes_2_13_2024.csv'
uv_vis = "/content/drive/Shared drives/Olshansky Lab/Data/UV_Vis/26' Yurii Sulima/" + name
uv_data1 = np.genfromtxt(uv_vis, skip_header = 2, skip_footer = 70, delimiter = ',', dtype = float)
data1 = uv_data1[:,:4]
data1 = data1.astype(np.double)
wavelengthes = data1[:,0]
intensities1 = data1[:,3]
#print(wavelengthes)
```

```python
maxIntensity, maxWavelength = 0,0
for index in range(0,len(wavelengthes)-1):
  if wavelengthes[index] > 380 and wavelengthes[index] < 450:
    if maxIntensity < intensities1[index]:
      maxIntensity = intensities1[index]
      maxWavelength = wavelengthes[index]
print(maxIntensity)
print(maxWavelength)
```

```python
font = {'family':'serif','color':'darkred','size':15}
fig, pointer = plt.subplots()
plt.xlabel('Wavelength (nm)', fontsize = 15, fontdict = font)
plt.ylabel('Absorbance', fontsize = 15, fontdict = font)
plt.title("YS01 ZnSe in MilliQ 2/13/2024",fontsize = 15, fontdict = font)
plt.xlim(300,700)
plt.ylim(0,0.35)
plt.annotate('Max Intensity: ' + str(maxIntensity.round(4)) + "     Corresponding Wavelength: " + str(maxWavelength.round(2)),
            xy=(50, 3), xycoords='figure pixels')
line = plt.plot(wavelengthes,intensities1, label = "YS01", color='blue')


plt.grid()
plt.legend()
plt.show()
```
